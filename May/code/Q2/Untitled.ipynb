```python
event_num=df_combine_2.groupby('TypeText').count()['NOPD_Item']
```

```python
event_index=list(event_num[event_num>100].index)
```

```python
df_combine_2['>100']=df_combine_2['TypeText'].apply(lambda x: 1 if x in event_index else 0)
```

```python
df_combine_4=df_combine_2[df_combine_2['>100']==1]
```

```python
df_combine_5=df_combine_4[df_combine_4['PoliceDistrict']!=0]
```

```python
prob_all=df_combine_5.groupby('TypeText').count()['NOPD_Item']/df_combine_5.shape[0]*100
prob_all
```

```python
def cal_prob(n):
    """
    This function calculate the probability of each type of events of a data set.
    """
    df=df_combine_5[df_combine_5['PoliceDistrict']==n]
    prob=df.groupby('TypeText').count()['NOPD_Item']/df.shape[0]*100
    
    return prob
```

```python
result=[]
for i in range(1,9):
    prob=cal_prob(i)
    result.append((prob/prob_all).sort_values(ascending=False)[0])
print(np.max(result))
```

```python
df_2020=pd.read_csv('Call_for_Service_2020.csv')
df_2019=pd.read_csv('Calls_for_Service_2019.csv')
df_2018=pd.read_csv('Calls_for_Service_2018.csv')
df_2017=pd.read_csv('Calls_for_Service_2017.csv')
df_2016=pd.read_csv('Calls_for_Service_2016.csv')
```

```python
df_2020['Location']=df_2020['Location'].apply(lambda x: x[7:-1])
df_2020['Location']=df_2020['Location'].apply(lambda x: x.split(' '))
```

```python
df_2019=df_2019[~df_2019['Location'].isnull()]
df_2018=df_2018[~df_2018['Location'].isnull()]
df_2017=df_2017[~df_2017['Location'].isnull()]
df_2016=df_2016[~df_2016['Location'].isnull()]
```

```python
df_2019['Location']=df_2019['Location'].apply(lambda x: x[1:-1].split(', '))
df_2018['Location']=df_2018['Location'].apply(lambda x: x[1:-1].split(', '))
df_2017['Location']=df_2017['Location'].apply(lambda x: x[1:-1].split(', '))
df_2016['Location']=df_2016['Location'].apply(lambda x: x[1:-1].split(', '))
```

```python
df_2020=df_2020.rename(columns={'Type':'Type_'})
df_2019=df_2019.rename(columns={'Type':'Type_','TimeArrival':'TimeArrive'})
df_combine_6=pd.concat([df_2016,df_2017,df_2018,df_2019,df_2020])
```

```python
df_combine_6.sample(50)
```

```python
def to_float(x):
    try:
        return np.round(float(x[0]),5)
    except:
        return np.nan
```

```python
def to_float_2(x):
    try:
        return np.round(float(x[1]),5)
    except:
        return np.nan
```

```python
df_combine_6['X']=df_combine_6['Location'].apply(lambda x: to_float(x))
```

```python
df_combine_6['Y']=df_combine_6['Location'].apply(lambda x: to_float_2(x))
```

```python
df_combine_6=df_combine_6[~df_combine_6['X'].isnull()]
```

```python
df_combine_6=df_combine_6[~df_combine_6['Y'].isnull()]
```

```python
df_combine_6=df_combine_6[df_combine_6['X']!=0.00000]
```

```python
df_combine_6=df_combine_6[df_combine_6['Y']!=0.00000]
```

```python
df_combine_7=df_combine_6[df_combine_6['X']>0]
```

```python
df_combine_8=df_combine_6[df_combine_6['X']<0]
```

```python
df_combine_8=df_combine_8.rename(columns={'X':'Y','Y':'X'})
```

```python
df_final=pd.concat([df_combine_7,df_combine_8])
```

```python
df_final.sample(50)
```

```python
def cal_area(df,n):
    """
    This function calculate the area of a district
    """
    df=df[df['PoliceDistrict']==n]
    a=df['X'].std()*111
    b=np.abs(df['Y'].std())*111
    result=np.pi*a*b
    
    return result
```

```python
for i in range(1,9):
    print(cal_area(df_final,i))
```
