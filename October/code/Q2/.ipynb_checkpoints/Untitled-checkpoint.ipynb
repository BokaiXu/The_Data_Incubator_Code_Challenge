```python
# Q7: How far must one travel from one hotspot to another?
```

```python
df1.sample(5)
```

```python
df1.describe()
```

```python
def cal_distance(lat1,long1,lat2,long2):
    lat_m=(lat1+lat2)/2
    delta_lat=lat1-lat2
    delta_long=long1-long2
    R=6371
    distance=R*np.sqrt(delta_lat*delta_lat+(np.cos(lat_m)*delta_long)**2)
    distance_feet=3280.8399*distance
    return distance_feet
```

```python
def radians(degree):
    return degree*np.pi/180
```

```python
df1.loc[:,'long']=df1['Longitude'].apply(lambda x: radians(x))
```

```python
df1.loc[:,'lat']=df1['Latitude'].apply(lambda x: radians(x))
```

```python
df1.sample(5)
```

```python
from collections import defaultdict
distance_dict=defaultdict()
for i in df1['OBJECTID'].values:
     distance_dict.update({i:list()})
```

```python
for i in df1['OBJECTID'].values:
    for j in df1['OBJECTID'].values:
        lat1=df1[df1['OBJECTID']==i]['lat'].values[0]
        long1=df1[df1['OBJECTID']==i]['long'].values[0]
        lat2=df1[df1['OBJECTID']==j]['lat'].values[0]
        long2=df1[df1['OBJECTID']==j]['long'].values[0]
        distance=cal_distance(lat1,long1,lat2,long2)
        distance_dict[i].append(distance)
```

```python
df_distance=pd.DataFrame(distance_dict)
```

```python
df_distance
```

```python
distance_list=[]
for i in df_distance.columns:
    distance_list.append(df_distance[i].sort_values()[:4].sum()/3)
```

```python
np.median(np.array(distance_list))
```

```python
# Q8: Using a linear estimate for the number of monthly activations, what is rate of increase in monthly activations?
```

```python
df1=pd.read_csv('NYC_Wi-Fi_Hotspot_Locations.csv')
```

```python
df1.loc[:,'year']=df1['Activated'].str[-4:]
df1
```

```python
df1_new=df1[df1['year']!='9999']
```

```python
df1_new['month']=pd.to_datetime(df1_new['Activated']).dt.month
```

```python
df1_new
```

```python
df_month=df1_new.groupby(['year','month']).count()['Activated'].reset_index()
```

```python
df_month['aggre_month']=range(1,40)
```

```python
df_month.plot(x='aggre_month',y='Activated',kind='scatter')
```

```python
df_month
```

```python
df_month_linear=df_month.loc[:29]
```

```python
df_month_linear
```

```python
df_month_linear
```

```python
import statsmodels.api as sm
df_month_linear['intercept']=1
mod = sm.OLS(df_month_linear['Activated'], df_month_linear[['intercept','aggre_month']])
results = mod.fit()
results.summary()
```
